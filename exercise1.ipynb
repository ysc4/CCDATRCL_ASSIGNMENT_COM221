## Class Initialization

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class LinkedList:

    def __init__(self):
        self.head = None

    def traverse(self, node):
        while node:
            print(node.data, end=" -> ")
            node = node.next
```

```python
class DoublyLinkedList:

    def __init__(self):
        self.head = None

    def traverse(self, node):
        while node:
            print(node.data, end=" -> ")
            node = node.next
```

```python
class CircularLinkedList:

    def __init__(self):
        self.head = None

    def traverse(self, node):

        head = node

        while node:
            print(node.data, end=" -> ")
            node = node.next

            # Break to prevent infinite loop
            if node == head:
                break
```

#### 1. Create a Singly Linked list

- Each node should contain your name and your seatmates in your row
- Create at least 5 nodes
- Traverse all nodes

```python
singly = LinkedList()

singly.head = Node('Kalli')
node_b = Node('Boybie')
node_c = Node('Jessy')
node_d = Node('Jassy')
node_e = Node('Angelo')
node_f = Node('Dhan')
node_g = Node('Robbie')
node_h = Node('Iber')

singly.head.next = node_b
node_b.next = node_c
node_c.next = node_d
node_d.next = node_e
node_e.next = node_f
node_f.next = node_g
node_g.next = node_h
node_h.next = None

singly.traverse(singly.head)
```

#### 2. Create a Doubly Linked list

- Each node should contain your name and your seatmates in your row
- Create at least 5 nodes
- Traverse all nodes

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.prev = None

doubly = DoublyLinkedList()

doubly.head = Node('Kalli')
node_b = Node('Boybie')
node_c = Node('Jessy')
node_d = Node('Jassy')
node_e = Node('Angelo')
node_f = Node('Dhan')
node_g = Node('Robbie')
node_h = Node('Iber')

doubly.head.prev = None
doubly.head.next = node_b

node_b.prev = doubly.head
node_b.next = node_c

node_c.prev = node_b
node_c.next = node_d

node_d.prev = node_c
node_d.next = node_e

node_e.prev = node_d
node_e.next = node_f

node_f.prev = node_e
node_f.next = node_g

node_g.prev = node_f
node_g.next = node_h

node_h.prev = node_g
node_h.next = None

doubly.traverse(doubly.head)
```

#### 3. Create a Circular Linked list

- Each node should contain your name and your seatmates in your row
- Create at least 5 nodes
- Traverse all nodes

```python
circular = CircularLinkedList()

circular.head = Node('Kalli')
node_b = Node('Boybie')
node_c = Node('Jessy')
node_d = Node('Jassy')
node_e = Node('Angelo')
node_f = Node('Dhan')
node_g = Node('Robbie')
node_h = Node('Iber')

circular.head.next = node_b
node_b.next = node_c
node_c.next = node_d
node_d.next = node_e
node_e.next = node_f
node_f.next = node_g
node_g.next = node_h
node_h.next = circular.head

circular.traverse(circular.head)
```
